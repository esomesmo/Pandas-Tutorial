# How to Change the Order of Columns in Pandas DataFrame

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/softhints/Pandas-Tutorials/master/data/population/population.csv')
df
```

```python
df.columns
```

```python
import random

initial_order = df.columns.to_list()
cols = df.columns.to_list()

random.shuffle(cols)
df
```

```python
df = df.reindex(cols, axis=1)
df
```

```python
df.columns
```

```python
df = df.reindex(sorted(df.columns), axis=1)
df
```

```python
df = df.sort_index(axis=1)
df
```

```python
df[sorted(df.columns)]
```

```python
df.columns
```

## Shift columns

```python
df
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/softhints/Pandas-Tutorials/master/data/population/population.csv')
cols = df.columns.tolist()
cols = cols[-2:] + cols[:-2]
cols
```

```python
print(cols)
```

```python
df = df.reindex(cols, axis=1)
df
```

```python
df = df[cols]
df
```

## Performance

```python
%timeit df.sort_index(axis=1)
```

```python
df.columns = initial_order
```

```python
%timeit df.reindex(sorted(initial_order), axis=1)
```

```python
df_m = pd.concat([df] * 100000, axis=0)
df_m.shape
```

```python
%timeit df_m.sort_index(axis=1)
```

```python
%timeit df_m.reindex(sorted(df_m.columns), axis=1)
```

```python
df_m = pd.concat([df] * 100, axis=1)
df_m.shape
```

```python
c = df_m.T.reset_index(drop=True).T.columns.to_list()
random.shuffle(c)
df_m.columns = c
```

```python
%timeit df_m.sort_index(axis=1)
```

```python
c = df_m.T.reset_index(drop=True).T.columns.to_list()
random.shuffle(c)
df_m.columns = c
```

```python
%timeit df_m.reindex(sorted(c), axis=1)
```

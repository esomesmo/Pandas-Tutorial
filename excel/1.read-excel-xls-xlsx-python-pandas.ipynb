````python
import pandas as pd

pd.read_excel('animals.xlsx')
````

````python
pd.read_excel('animals.xlsx', sheet_name="Sheet2")
````

````python
pd.read_excel('animals.xlsx', sheet_name=["Sheet1", "Sheet2"])
````

````python
pd.read_excel('animals.xlsx', sheet_name=["Sheet1", "Sheet2"]).get('Sheet1')
````

````python
pd.read_excel('animals.xlsx', skiprows=2, nrows=3)
````

````python
pd.read_excel('animals.xlsx', usecols='C:D')
````

````python
pd.read_excel('animals.xlsx', skiprows=2, nrows=3, usecols='C:D')
````

````python
pd.read_excel('animals.xlsx', index_col=2)
````

````python
pd.read_excel('animals.xlsx', usecols="A:C")
````

````python
pd.read_excel('animals.xlsx', index_col=2)
````

````python
pd.read_excel('animals.xlsx', engine='xlrd')
````

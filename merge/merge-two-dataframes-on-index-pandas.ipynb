```python
How to Merge two dataframes by index IN pandas
```

```python
import pandas as pd
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)
display(df1)


df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[0, 1, 2, 2],
)
```

## Step 1: Pandas: merge on index by method `merge`

```python
df_m = pd.merge(df1, df2, left_index=True, right_index=True)
df_m
```

```python
df_m = pd.merge(df1, df2, left_index=True, right_index=True, suffixes=('_left', '_right'))
df_m
```

```python
import pandas as pd
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 2],
)
display(df1)


df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[0, 1, 2, 2],
)
display(df2)
df_m = pd.merge(df1, df2, left_index=True, right_index=True)
df_m
```

## Step 2: Pandas: merge on index by `concat` and axis=1

```python
df_m = pd.concat([df1, df2], axis='columns')
df_m
```

```python
df_m['A']
```

```python
import pandas as pd
import traceback

try:
    
    df_m = pd.concat([df1, df2], axis='columns', verify_integrity=True)

except Exception as e:
    traceback.print_exc(limit=1)
    exit(1)
```

## Step 3: Pandas: merge on index - `join`

```python
df1.join(df2, lsuffix='_x')
```

## Step 4: Pandas: merge on index - `merge` vs `concat`

```python
import pandas as pd
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)


df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7", "A8"],
        "B": ["B4", "B5", "B6", "B7", "B8"],
        "C": ["C4", "C5", "C6", "C7", "C8"],
        "D": ["D4", "D5", "D6", "D7", "D8"],
    },
    index=[0, 1, 2, 3, 4],
)
```

```python
df_m = pd.merge(df1, df2, left_index=True, right_index=True)
df_m
```

```python
df_m = pd.merge(df1, df2, left_index=True, right_index=True, how='outer')
df_m
```

```python
df_m = pd.concat([df1, df2], axis='columns')
df_m
```

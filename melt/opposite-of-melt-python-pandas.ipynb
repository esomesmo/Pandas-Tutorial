# Opposite of Melt in Python and Pandas

```python
import pandas as pd

df_pop = pd.read_csv('https://raw.githubusercontent.com/softhints/Pandas-Tutorials/master/data/population/population.csv')
df_pop
```

```python
df_m = df_pop.melt(id_vars=['Region'])
df_m
```

```python
df_m.pivot(*df_m).reset_index()
```

```python
df_m.pivot(*df_m).reset_index().columns
```

```python
df_m.pivot(*df_m).reset_index().rename_axis(None, axis='columns')
```

```python
df_m.pivot(*df_m).reset_index().rename_axis(None, axis='columns').columns
```

## Melt on few columns

```python
df_m = df_pop.melt(id_vars=['Region'], value_vars=['1500', '1600', '1700'])
df_m
```

```python
import pyperclip

pyperclip.copy((df_m.head().to_html(classes='table table-striped text-center', justify='center', index=False)))
```

```python
df_m.pivot(index='Region', columns='variable')['value']
```

```python
import pyperclip

pyperclip.copy((df_m.pivot(index='Region', columns='variable')['value'].head().to_html(classes='table table-striped text-center', justify='center')))
```
